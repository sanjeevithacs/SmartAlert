```python
from datetime import datetime, timedelta

deadline = datetime(2025, 8, 11, 10, 30)  # YYYY, M, D, H, M
now = datetime.now()

if now >= deadline - timedelta(minutes=10) and now < deadline:
    reminder_type = "before_deadline"
elif now >= deadline:
    reminder_type = "after_deadline"
else:
    reminder_type = None


print(f"🕒 Current time: {now}")
print(f"📌 Reminder Type: {reminder_type}")
```

```python
import smtplib
from email.mime.text import MIMEText

def send_email(to, subject, body):
    msg = MIMEText(body)
    msg['From'] = "you@example.com"
    msg['To'] = to
    msg['Subject'] = subject

    server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
    server.login("you@example.com", "your_password")
    server.sendmail("you@example.com", to, msg.as_string())
    server.quit()

if reminder_type:
    for _, row in non_responders.iterrows():
        if reminder_type == "before_deadline":
            print(f"📧 Sending reminder to {row['Name']} ({row['Email']}): Fill the form within 10 minutes!")
        elif reminder_type == "after_deadline":
            print(f"📧 Sending message to {row['Name']} ({row['Email']}): Deadline has passed!")
else:
    print("⏳ Not time to send reminders yet.")

```
